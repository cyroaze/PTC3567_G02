```python
from matplotlib import pyplot as plt
import pandas as pd
from datetime import datetime as dt
import seaborn as sns
import statsmodels.api as sm
import numpy as np
```

```python
dfs = []
total_rows = 0  # Variável para armazenar o número total de linhas

for year in [2019, 2020, 2021, 2022]:
    for semester in [1, 2]:
        file_path = f'Data/ca-{year}-0{semester}.csv'
        temp_df = pd.read_csv(file_path, delimiter=';', decimal=',', encoding='utf8')
        num_rows = len(temp_df)  # Conta o número de linhas no DataFrame
        total_rows += num_rows
        print(f'O arquivo {file_path} tem {num_rows} linhas.')
        dfs.append(temp_df)

df = pd.concat(dfs, ignore_index=True)

# Imprime o número de linhas do DataFrame final
print(f'O DataFrame final tem {total_rows} linhas.')

# Cria um arquivo CSV com o DataFrame resultante
df.to_csv('Dataset_preco_combustiveis.csv', index=False, sep=';', decimal=',', encoding='utf-8')
```
